```python
import json
import pickle 
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = '.\\19_23_func1.json'
df_1 = pd.read_json(file_path)
b = pd.json_normalize(df_1['_source'],max_level=1)
type(df_1), type(b)
```

```python
b
```

```python
HR_data15 = {}
HR_time15 = {}
tmp = dict(b['layers.modbus'])
tmp1 = dict(b["layers.frame"])

for j in range(3601):
     for i in tmp[j].keys():
        
        time = tmp1[j]['frame.time_epoch']
        if "Bit " in i:
            #'Bit 1001 : 0'
            num=i[4:i.find(" :")]
            val=int(i[i.find(": ")+2:])
        else:
            continue
        if num not in HR_data15.keys():
            HR_data15[num]=list()
            HR_time15[num]=list()
        HR_data15[num].append(val)
        HR_time15[num].append(time)
HR_data15.keys()
```

```python
for i in range(15):
    plt.plot(np.array(HR_time15[str(i+3)], dtype='float')-1492392300.740706000,HR_data15[str(i+3)], label=i+3, color='b')
    
    plt.legend()
    plt.ylim(-1,2)
    plt.show()
```

## function 5 with 23-55

```python
file_path = '.\\55_23_func5.json'
with open(file_path, encoding='utf-8') as f:
    pkts=json.load(f)
df_3 = pd.DataFrame(pkts)
b = pd.json_normalize(df_3['_source'],max_level=1)
b
```

```python
valList=[]
HR_data = {}
HR_time = {}
tmp = dict(b['layers.modbus'])
tmp1 = dict(b["layers.frame"])

for j in range(100):
    status = tmp[j]['modbus.data']
    addr = tmp[j]['modbus.reference_num']
    time = tmp1[j]['frame.time_epoch']
    if status == 'ff:00':
        val=1
    elif status == '00:00':
        val=0
    
    if addr not in HR_data.keys():
        HR_data[addr]=list()
        HR_time[addr]=list()
    HR_data[addr].append(val)
    HR_time[addr].append(time)
HR_data.keys()
```

```python
# plt.plot(HR_time['1003'],HR_data['1003'], label='Register1003')
# plt.plot(HR_time['1001'],HR_data['1001'], label='Register1001')
plt.scatter(np.array(HR_time['1000'], dtype='float')-1492392300.740706000,HR_data['1000'], label='Register1000')
plt.legend()
plt.title('Holding Registers 1000, 1001, 1003')
```
