```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import transforms
import matplotlib.cm as cm
import cv2
```

```python
# file_path = "HectorDebugCost.csv"
with open('HectorDebugCost.csv', 'r') as csvfile:
    data = list( csv.reader(csvfile, delimiter=',') )
    timestamp = np.array(data)[:,0].astype('float32') # dtype: U5 means string length < 5
    cost1 = np.array(data)[:,1].astype('float32')
    cost2 = np.array(data)[:,2].astype('float32')
    cost3 = np.array(data)[:,3].astype('float32')
    
    flag1 = np.array(data)[:,4].astype('uint8')
    flag2 = np.array(data)[:,5].astype('uint8')
    flag3 = np.array(data)[:,6].astype('uint8')
```

```python
fig, ax = plt.subplots(figsize=(12,3), dpi=200)
plt.plot(cost1, color='b', label='scale1', linewidth=0.3)
plt.plot(cost2, color='r', label='scale2', linewidth=0.3)
plt.plot(cost3, color='tab:green', label='scale3', linewidth=0.3)
plt.legend()
plt.title('cost of different scale')
```
